# Generating and Manipulating ndarrays

1) np.ones(): It is used to create an array of 1s.

2) np.zeros(): It is used to create an array of 0s.

3) np.random.randint(): It is used to create a random array of integers within a particular range.

4) np.random.random(): It is used to create an array of random numbers.

5) np.arange(): It is used to create an array with increments of fixed step size.

6) np.linspace(): It is used to create an array of fixed length.

```python
# Use .arange() to create a vector with values ranging from 0 to n-1.
a = np.arange(1,11)
a
```

```python
# Slicing--> last 3 values
a[-3:]
```

```python
# subset with only 4th, 5th and 6th values
a[3:6]
```

```python
# subset having only even integers
a[a % 2== 0]
```

```python
# Reverse the order of the values in a 
np.flip(a)
```

```python
a[::-1]
```

### Matrix Multiplication

```python
# Matrix Multiplication
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])
 
b1 = np.array([[4, 1, 3],
               [5, 8, 5]])
```

```python
a1.shape
```

```python
b1.shape
```

```python
# Matrix multiplication only possible when column of a1 == row of b1 and shape of resultant will be row of a1 x column of b1
c = np.matmul(a1, b1)
c
```

```python
c.shape
```

```python
# b1 x a1 is not possible --> (2,3) x (4,2) --> 3 != 4 thats why its not posssible
np.matmul(b1,a1)
```

# Manipulating Images as ndarrays

```python
import matplotlib.pyplot as plt
from scipy import misc # contains an image of a racoon!
from PIL import Image # for reading image files
```

```python
img = misc.face()
plt.imshow(img)
```

```python
type(img)
```

```python
img.shape
```

There are three matrices stacked on top of each other - one for the red values, one for the green values and one for the blue values. Each matrix has a 768 rows and 1024 columns, which makes sense since 768x1024 is the resolution of the image.

Now can we try and convert the image to black and white? All we need need to do is use a formula.
Y_linear = 0.2126R + 0.7152G + 0.0722B



### Converting an image to grayscale

```python
# 1st step--> converting as values to RGBs by dividing it by 255
sRGB_array = img / 255
```

```python
# 2nd step--> creating gray_vals array
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
# 3rd step --> matrix multiplication
img_gray = sRGB_array @ grey_vals
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
# The cmap parameter is important here. It will not know that is dealing with a black and white image.
plt.imshow(img_gray)
```
